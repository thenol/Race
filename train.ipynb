```python
# training & checking

df_f = pd.read_csv('features.csv')
a = df_f[df_f.duplicated('uId')]
```

```python

l = list(a.index)
df_u = df_f.drop(l)

df_u
```

```python
train = pd.read_csv('age_train.csv',names=['uId','label'])
```

```python
df_data=pd.merge(df_u,train)
# df_data.shape
# df_data
def appId_to_int(r):
    r['app_top']=int(r['app_top'][1:])
    return r
df_data=df_data.apply(appId_to_int,axis=1)
```

```python
def dec(r):
    r['label']-=1
    return r
df_data = df_data.apply(dec,axis=1)
```

```python
import json
import lightgbm as lgb
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.datasets import  make_classification
 
# iris = load_iris()   # è½½å…¥é¸¢å°¾èŠ±æ•°æ®é›†
# data=iris.data
# target = iris.target

data = df_data.values[:,0:20]
target = df_data.values[:,20]

X_train,X_test,y_train,y_test =train_test_split(data,target,test_size=0.2)
 
 
# åŠ è½½ä½ çš„æ•°æ®
# print('Load data...')
# df_train = pd.read_csv('../regression/regression.train', header=None, sep='\t')
# df_test = pd.read_csv('../regression/regression.test', header=None, sep='\t')
#
# y_train = df_train[0].values
# y_test = df_test[0].values
# X_train = df_train.drop(0, axis=1).values
# X_test = df_test.drop(0, axis=1).values
 
# åˆ›å»ºæˆlgbç‰¹å¾çš„æ•°æ®é›†æ ¼å¼
lgb_train = lgb.Dataset(X_train, y_train) # å°†æ•°æ®ä¿å­˜åˆ°LightGBMäºŒè¿›åˆ¶æ–‡ä»¶å°†ä½¿åŠ è½½æ›´å¿«
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)  # åˆ›å»ºéªŒè¯æ•°æ®
 
# å°†å‚æ•°å†™æˆå­—å…¸ä¸‹å½¢å¼
params = {
    'learning_rate':0.1,
    'lambda_l1':0.1,
    'lambda_l2':0.2,
    'max_depth':4,
    'objective':'multiclass',
    'num_class':6,  #lightgbm.basic.LightGBMError: bâ€˜Number of classes should be specified and greater than 1 for multiclass trainingâ€˜
}
 
print('Start training...')
# è®­ç»ƒ cv and train
gbm = lgb.train(params,lgb_train,num_boost_round=20,valid_sets=lgb_eval,early_stopping_rounds=5) # è®­ç»ƒæ•°æ®éœ€è¦å‚æ•°åˆ—è¡¨å’Œæ•°æ®é›†
 
# print('Save model...') 
 
# gbm.save_model('model.txt')   # è®­ç»ƒåŽä¿å­˜æ¨¡åž‹åˆ°æ–‡ä»¶
 
print('Start predicting...')
# é¢„æµ‹æ•°æ®é›†
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration) #å¦‚æžœåœ¨è®­ç»ƒæœŸé—´å¯ç”¨äº†æ—©æœŸåœæ­¢ï¼Œå¯ä»¥é€šè¿‡best_iterationæ–¹å¼ä»Žæœ€ä½³è¿­ä»£ä¸­èŽ·å¾—é¢„æµ‹
# è¯„ä¼°æ¨¡åž‹
print('The rmse of prediction is:', mean_squared_error(y_test, y_pred) ** 0.5) # è®¡ç®—çœŸå®žå€¼å’Œé¢„æµ‹å€¼ä¹‹é—´çš„å‡æ–¹æ ¹è¯¯å·®
```

```python
import lightgbm as lgb
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,accuracy_score


data = df_data.values[:,1:20]
label = df_data.values[:,20]
X_train,X_test,y_train,y_test=train_test_split(data,label,test_size=0.2)

train_data=lgb.Dataset(X_train,label=y_train)
validation_data=lgb.Dataset(X_test,label=y_test)
params={
    'learning_rate':0.8,
    'lambda_l1':0,
    'lambda_l2':0,
    'max_depth':6,
    'objective':'multiclass',
    'num_class':6,
    'num_leaves': 31,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
 
    'num_threads':4
}
clf=lgb.train(params,train_data,valid_sets=[train_data,validation_data],num_boost_round=100,early_stopping_rounds=10)

y_pred=clf.predict(X_test)
y_pred=np.argmax(y_pred,axis=1)
print(accuracy_score(y_test,y_pred))
```
